```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
#graph_type = 'Random'
#graph_type = 'Grid'
#graph_type = 'Cycle'
#graph_type = 'Barbell'
#graph_type = 'Complete'
graph_type = 'Star'
#graph_type = 'Turan'
#graph_type = 'Wheel'
#graph_type = 'ER'
#graph_type = '4Reg'
```

```python
list_cost_0 = []
list_cost_1 = []
list_cost_2 = []
list_cost_3 = []
list_cost_4 = []

with open('Data/QGK_' + graph_type + '_List_Cost0.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        list_cost_0.append(row)
    list_cost_0 = list_cost_0[1]  
    for i in range(len(list_cost_0)):
        list_cost_0[i] = float(list_cost_0[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_List_Cost1.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        list_cost_1.append(row)
    list_cost_1 = list_cost_1[1]  
    for i in range(len(list_cost_1)):
        list_cost_1[i] = float(list_cost_1[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_List_Cost2.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        list_cost_2.append(row)
    list_cost_2 = list_cost_2[1]  
    for i in range(len(list_cost_2)):
        list_cost_2[i] = float(list_cost_2[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_List_Cost3.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        list_cost_3.append(row)
    list_cost_3 = list_cost_3[1]  
    for i in range(len(list_cost_3)):
        list_cost_3[i] = float(list_cost_3[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_List_Cost4.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        list_cost_4.append(row)
    list_cost_4 = list_cost_4[1]  
    for i in range(len(list_cost_4)):
        list_cost_4[i] = float(list_cost_4[i])  
    csvdatei1.close()
```

```python
sandwich_cost_0 = []
sandwich_cost_1 = []
sandwich_cost_2 = []
sandwich_cost_3 = []
sandwich_cost_4 = []

with open('Data/QGK_' + graph_type + '_Sandwich_Cost0.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        sandwich_cost_0.append(row)
    sandwich_cost_0 = sandwich_cost_0[0]  
    for i in range(len(sandwich_cost_0)):
        sandwich_cost_0[i] = float(sandwich_cost_0[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_Sandwich_Cost1.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        sandwich_cost_1.append(row)
    sandwich_cost_1 = sandwich_cost_1[0]  
    for i in range(len(sandwich_cost_1)):
        sandwich_cost_1[i] = float(sandwich_cost_1[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_Sandwich_Cost2.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        sandwich_cost_2.append(row)
    sandwich_cost_2 = sandwich_cost_2[0]  
    for i in range(len(sandwich_cost_2)):
        sandwich_cost_2[i] = float(sandwich_cost_2[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_Sandwich_Cost3.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        sandwich_cost_3.append(row)
    sandwich_cost_3 = sandwich_cost_3[0]  
    for i in range(len(sandwich_cost_3)):
        sandwich_cost_3[i] = float(sandwich_cost_3[i])  
    csvdatei1.close()

with open('Data/QGK_' + graph_type + '_Sandwich_Cost4.csv') as csvdatei1:
    csv_reader_object = csv.reader(csvdatei1)
    for row in csv_reader_object:
        sandwich_cost_4.append(row)
    sandwich_cost_4 = sandwich_cost_4[0]  
    for i in range(len(sandwich_cost_4)):
        sandwich_cost_4[i] = float(sandwich_cost_4[i])  
    csvdatei1.close()
```

```python
list_mean_cost = []
list_std_cost = []
list_std_upper_cost = []
list_std_lower_cost = []
for i in range(len(list_cost_0)):
    list_mean_cost.append(np.mean([list_cost_0[i], list_cost_1[i], list_cost_2[i], list_cost_3[i], list_cost_4[i]]))
    list_std_cost.append(np.std([list_cost_0[i], list_cost_1[i], list_cost_2[i], list_cost_3[i], list_cost_4[i]]))

for i in range(len(list_cost_0)):
    list_std_upper_cost.append(list_mean_cost[i] + list_std_cost[i])
    list_std_lower_cost.append(list_mean_cost[i] - list_std_cost[i])
```

```python
sandwich_mean_cost = []
sandwich_std_cost = []
sandwich_std_upper_cost = []
sandwich_std_lower_cost = []
for i in range(len(sandwich_cost_0)):
    sandwich_mean_cost.append(np.mean([sandwich_cost_0[i], sandwich_cost_1[i], sandwich_cost_2[i], sandwich_cost_3[i], sandwich_cost_4[i]]))
    sandwich_std_cost.append(np.std([sandwich_cost_0[i], sandwich_cost_1[i], sandwich_cost_2[i], sandwich_cost_3[i], sandwich_cost_4[i]]))

for i in range(len(sandwich_cost_0)):
    sandwich_std_upper_cost.append(sandwich_mean_cost[i] + sandwich_std_cost[i])
    sandwich_std_lower_cost.append(sandwich_mean_cost[i] - sandwich_std_cost[i])
```

```python
print(list_mean_cost)
print(list_std_cost)
print(list_std_lower_cost)
```

```python
t = np.arange(3001)
plt.figure()

plt.plot(list_mean_cost, label='list')
plt.plot(sandwich_mean_cost, label='sandwich')
plt.legend(loc = 'upper right')
plt.title('Quantum Graph Kernel - ' + graph_type)
ax = plt.gca()
#ax.set_ylim([0.0, 1])
ax.fill_between(t, list_std_upper_cost, list_std_lower_cost, alpha = 0.5)
ax.fill_between(t, sandwich_std_upper_cost, sandwich_std_lower_cost, alpha = 0.5)
plt.xlabel("Step")
plt.ylabel("costs")

plt.savefig('Graphics_Training/QGK_' + graph_type +'.svg', format='svg', dpi = 1200)
```
